<h3><em><b>3.1.7 Regresión entre precio de mercado y atributos.</b></em></h3>
-------------------

Objetivo:
- Se busca encontrar una relación entre el precio de mercado que possen los jugadores en el juego y las valoraciones que presentar referidas a algunas habilidades principales.
- Con esto, intentamos determinar cuales son las habilidades mejor valoradas por el mercado.

```python
# Creamos el modelo de regresión para los atributos y valores de mercado de los jugadores
# Creamos variables a comparar
X = Regresion[['Value']]
y_ofens = Regresion[['Agility', 'Dribbling', 'Finishing', 'Shot power','Sprint speed', 'Interceptions', 'Marking']]

# Creamos una figura con subgráficas
plt.figure(figsize=(15, 12))

for i, col in enumerate(y_ofens.columns):
    y_of = Regresion[col]
    
    # Creamos y ajustamos el modelo de regresión lineal
    model = LinearRegression()
    model.fit(X, y_of)
    y_pred_of = model.predict(X)
    
    # Creamos una subgráfica para cada atributo
    plt.subplot(4, 2, i + 1)
    plt.scatter(Regresion['Value'], y_of, label=f'Datos {col}', color='blue', alpha=0.6)
    plt.plot(Regresion['Value'], y_pred_of, label=f'Regresión {col}', color='red')
    
    # Configuramos el gráfico
    plt.xlabel('Valor de mercado [€]')
    plt.ylabel(col)
    plt.title(f'Regresión de {col}')
    plt.legend()
    plt.grid()

# Ajustamos el layout y mostramos la gráfica
plt.tight_layout()
plt.show()
```

Conclusión:
- Se observa una relación esperable creciente entre las valoraciones en las distintas habilidades y el precio del mercado. Ahora comparemos que tanto se incrementa el precio del mercado con el incremento de cada habilidad comparada.

<h3><em><b>3.1.8 Comparación entre las curvas de regresion para diferentes atributos</b></em></h3>
-------------------

```python
# Colores para cada habilidad
colores = {
    'Agility': 'blue',
    'Dribbling': 'orange',
    'Finishing': 'green',
    'Shot power': 'red',
    'Sprint speed': 'purple',
    'Interceptions': 'brown',
    'Marking': 'pink'
}

# Creamos la figura
plt.figure(figsize=(12, 8))

# Recorremos cada habilidad y ajustamos el modelo de regresión
for col in y_ofens.columns:
    y_of = Regresion[col]
    
    # Creamos y ajustamos el modelo de regresión lineal
    model = LinearRegression()
    model.fit(X, y_of)
    y_pred_of = model.predict(X)
    
    # Graficamos la regresión
    plt.plot(Regresion['Value'], y_pred_of, label=f'Regresión {col}', color=colores[col])

# Configuramos el gráfico
plt.xlabel('Valor de mercado [€]')
plt.ylabel('Habilidades')
plt.title('Curvas de Regresión de Habilidades vs Valor de Mercado')
plt.legend()
plt.grid()

# Mostramos la gráfica
plt.tight_layout()
plt.show()
```

Conclusión:
- Se observa que las habilidades cuyo incremento generan un mayor incremento en la valoración que el mercado realiza de los jugadores (pendiente de las rectas) son el regate, la finalización y el poder de remate. Estas habilidades tienen en común que generalmente se asocian a jugadores ofensivos.
- En un segundo grupo se observan la agilidad, la velocidad y la capacidad de intercepción.
- En un tercer grupo, como habilidad menos valorada se observa la capacidad de marca, ligada fundamentalmente a jugadores defensivos.
- Se conlcuye entonces que los jugadores mejores valorados en términos económicos por el mercado, son aquellos que destacan en habilidades ofensivas, como son delanteros y mediocampistas de ataque. En contrapartida, los defensores o mediocampistas asociados a habilidades defensivas, son menos valorados económicamente por el mercado.

```python
# Se obtienen los 5 mejores arqueros para la categoría "Joven Promesa".

mejores_arq_jov_idx = (Posiciones_arq[Posiciones_arq["Rango edad"] == "Joven promesa"])["GK"].sort_values(ascending=False)
mejores_arq_jov = Posiciones_arq.loc[mejores_arq_jov_idx.keys()].drop_duplicates().head(5)

# Se obtienen los 5 mejores arqueros para la categoría "Máximo potencial".

mejores_arq_maxpot_idx = (Posiciones_arq[Posiciones_arq["Rango edad"] == "Máximo potencial"])["GK"].sort_values(ascending=False)
mejores_arq_maxpot = Posiciones_arq.loc[mejores_arq_maxpot_idx.keys()].drop_duplicates().head(5)

# Se obtienen los 5 mejores arqueros para la categoría "Cercano al retiro".

mejores_arq_retiro_idx = (Posiciones_arq[Posiciones_arq["Rango edad"] == "Cercano al retiro"])["GK"].sort_values(ascending=False)
mejores_arq_retiro = Posiciones_arq.loc[mejores_arq_retiro_idx.keys()].drop_duplicates().head(5)
```

```python
# Gráficamos los mejores jugadores por posición para la categoria "Cercano al retiro"

plt.figure(figsize=(18,6))
sns.barplot(data=mejores_arq_jov, x=mejores_arq_jov["Name"], y=mejores_arq_jov["GK"], label="Joven Promesa")
sns.barplot(data=mejores_arq_maxpot, x=mejores_arq_maxpot["Name"], y=mejores_arq_maxpot["GK"], label="Máximo potencial")
sns.barplot(data=mejores_arq_retiro, x=mejores_arq_retiro["Name"], y=mejores_arq_retiro["GK"], label="Cercano al retiro")

# Refencias del gráfico

plt.title("Mejores jugadores según potencial por edad (arqueros)")
plt.xlabel("Nombre del jugador")
plt.ylabel("GK Score")
plt.tight_layout()

# Crear el gráfico
plt.show()
```
